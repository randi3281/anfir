```r
# Instal ggplot2 jika belum terinstal
install.packages("ggplot2")

# Muat ggplot2
library(ggplot2)

# Data
nama = c('budi', 'mey', 'refi', 'mega', 'jimmy')
uang = c(20000, 21111, 2222, 44444, 22221)
belajar = c(3,2,1,4,5)
df = data.frame(nama, uang, belajar)

# Plot dengan ggplot2
ggplot(df, aes(x = uang, y = belajar)) +
  geom_point() +
  theme(panel.background = element_rect(fill = "lightblue"))
```
