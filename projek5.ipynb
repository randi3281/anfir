# Simple Regresi Linear

```r
## Input Data ##
x <- c(40, 55, 32, 55, 50, 52, 61, 44, 30, 22, 40, 64, 58, 48, 44)
y <- c(4, 16, 12, 24, 15, 24, 22, 17, 4, 14, 24, 26, 20, 9, 14)

## Membuat scatter plot atau diagram pencar / sebaran ##
plot(x,y)

## Membuat garis regresi background putih, bukan garisnya yang putih ##
abline(lm(y~x))

lm(y~x) ## Melihat hasil regresi ##
```

# Summary Linear Regresi

```r
lm(formula = y ~ x)

summary(lm(y~x))
```
